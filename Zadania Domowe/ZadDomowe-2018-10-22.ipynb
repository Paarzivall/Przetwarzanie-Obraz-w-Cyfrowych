```python
from skimage import data, img_as_ubyte
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
```

```python
#importowanie obrazka
img = data.camera() 
```

```python
#wywietlenie obrazka
plt.figure(figsize=(5,5))
plt.imshow(img,cmap="gray")
plt.axis('off')
plt.show()
```

```python
def MyHistogram(img):
    img = img_as_ubyte(img)
    tabPikseli = [0] * 256
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            x = img[i][j]
            tabPikseli[x] += 1
    return tabPikseli
```

```python
#print(MyHistogram(img))
```

```python
OsX = range(256)#o X
OsY = MyHistogram(img)#o Y
#rysowanie histogramu
plt.title("Histogram stworzony przez wasn funkcj")
plt.bar(OsX, OsY)
plt.xlim([min(OsX)-10, max(OsX)+10])
plt.grid()
plt.show()
```

```python
plt.title("Histogram oryginalny")
histogram = exposure.histogram(img, nbins=256)
hist, cbins = histogram
plt.bar(cbins, hist)
plt.xlim([-10, 265])
plt.grid()
plt.show()
```
